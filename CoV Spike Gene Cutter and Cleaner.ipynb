Extraction Procedure of Spike Gene

```python
spike_genes = []
for record in SeqIO.parse (input_file , file_format):
    cutter = record [21500:25499]
    spike_genes.append (cutter)


spike_file = SeqIO.write (spike_genes,'SpikeFile.fasta', 'fasta')

   
#this for..loop is necessitating extraction of the approximate Spike gene for every sequence in the input file. The location is bigger than the actual locus of Spike gene in case given sequences do not have the same length
      
```

```python
for spike in SeqIO.parse("SpikeFile.fasta", 'fasta'):
    print(len(spike))            #length of the extracted Spike gene
    print(repr(spike.seq))
    print(spike.id)
    print((str(spike)).count("N"))       #if number of unidentified bases "N" is too high run the cleaning procedure
```


Sequence Cleaner
